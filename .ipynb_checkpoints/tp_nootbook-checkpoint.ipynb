## 3) Outliers en PREVAILING_WAGE (salario anual)

Como deciamos más arriba, la variable númerica que hace referencias del salario anual tiene un problema importante de contar con registros fuera de un rango esperable. 

Dijimos que encontramos valores de 0 salario y en los valores máximos de 297,785,280 ojala cobraramos eso. Para considerar como valores extremos obtubimos los percentiles 0.05 y 0.95, que nos muestra el mínimo y el máximo del 90% de la muestra. Para ser un poco más precisos, podemos pedir los percentiles 0.01 y el 0.99. Ahora lo que vamos a ver es que el 98% de los salarios de la muestra se moveran entre el 35,214 y en el extremo superior 183,061, dolares anuales.

Con los valores extremos podemos hacer lo siguiente:
* Eliminarlos de la muestra.
* Imputar los outliers.

Para nosotros la mejor decisión en este caso, es dejarlo como está. ¿Por qué? Porque depende de la aplicación posterior para la cual usemos los datos que podemos hacer. Si estamos estudiando fraudes, los extremos pueden ser casos interesantes para tener en la muestra.

```python
# Percentiles en salario anual
df_clean['PREVAILING_WAGE'].quantile(q= (0.01, 0.99) , interpolation='linear')
```

```python
# Boxplot de salario anual cortando la cola derecha de la distribución en el percentil 99  
per99 = df_clean['PREVAILING_WAGE'].quantile(q = 0.99, interpolation='linear')
data = df_clean.loc[df_clean['PREVAILING_WAGE'] <= per99, ]
fig, ax = plt.subplots(figsize = (14, 6))
fig.suptitle('Gráfico de caja: salario anual declarado, solicitudes de visa H1', fontsize=14)
g = sns.boxplot(x = 'PREVAILING_WAGE', data = data, ax=ax)
```

# GUARDAMOS NUESTRO MODELO DE DATOS EN UNA BASE DE DATOS

```python
# Conexion con pymsql
import pymysql
USER = 'userhost'
PASSWORD = '12345678'
SERVER = 'localhost'
PORT = '3306'
DB='visas_h1'
```

Antes de ejecutar los comandos de inserción, nosotros ejecutamos las consultas sql que se encuentran en el archivo **sql_queries.sql**. Allí lo que hacemos es crear la base de datos visa_h1, crear un nuevo usuario y darle privilegios, y finalmente creamos las tables según el diseño conceptual que venimos desarrollando. 

Para la inserción de datos creamos tres funciones una para cada tabla y finalmente las ejecutamos.

```python
# Insert datos en mysql
def insert_state(SERVER, USER, PASSWORD, DB):
    try:
        con = pymysql.connect( host=SERVER, user = USER, passwd=PASSWORD, db = DB )
        cursor = con.cursor()
        for index, row in codigos_state.iterrows():
            cursor.execute("""INSERT INTO attr_states (WORKSITE_STATE_ABB, WORKSITE_STATE_FULL) 
            values(%s,%s)""", [row.WORKSITE_STATE_ABB, row.WORKSITE_STATE_FULL])
        con.commit()
        cursor.close()
        con.close()
    except ValueError as e:
        print("Error ", e)
        con.close()
        exit(1)
    finally:
        if (con):
            con.close()
def insert_soc_code(SERVER, USER, PASSWORD, DB):
    try:
        con = pymysql.connect( host=SERVER, user = USER, passwd=PASSWORD, db = DB )
        cursor = con.cursor()
        for index, row in codigos_soc.iterrows():
            cursor.execute("""INSERT INTO attr_soc_names (SOC_CODE, SOC_NAME) 
            values(%s,%s)""", [row.SOC_CODE, row.SOC_NAME])
        con.commit()
        print('Registros insertados correctamente')
        cursor.close()
        con.close()
    except ValueError as e:
        print("Error ", e)
        con.close()
        exit(1)
    finally:
        if (con):
            con.close()
            
def insert_fact(SERVER, USER, PASSWORD, DB):
    try:
        con = pymysql.connect( host=SERVER, user = USER, passwd=PASSWORD, db = DB )
        cursor = con.cursor()
        for index, row in df_clean.iterrows():
            cursor.execute("""INSERT INTO fact_visas (CASE_NUMBER, CASE_STATUS, EMPLOYER_NAME, SOC_CODE, JOB_TITLE, FULL_TIME_POSITION, PREVAILING_WAGE, WORKSITE_CITY, YEAR, WORKSITE_STATE_FULL) 
            values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)""", [row.CASE_NUMBER, row.CASE_STATUS, row.EMPLOYER_NAME, row.SOC_CODE, row.JOB_TITLE, row.FULL_TIME_POSITION, row.PREVAILING_WAGE, row.WORKSITE_CITY, row.YEAR, row.WORKSITE_STATE_FULL])
        con.commit()
        print('Registros insertados correctamente')
        cursor.close()
        con.close()
    except ValueError as e:
        print("Error ", e)
        con.close()
        exit(1)
    finally:
        if (con):
            con.close()
```

```python
deseo = input('¿Queres guardar las tablas en la base de datos? y/n').lower()
if deseo=='y':
    #insert state
    insert_state(SERVER, USER, PASSWORD, DB)
    #insert soc
    insert_soc_code(SERVER, USER, PASSWORD, DB)
    #insert fact
    insert_fact(SERVER, USER, PASSWORD, DB)
```

**Listo!** Ya tenemos nuestros datos en una base de datos MySql, transformados, 'limpios' y ordenados.
